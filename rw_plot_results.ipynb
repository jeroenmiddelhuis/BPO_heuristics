```python
from rw_heuristics import random_policy, spt_policy, fifo_policy
from rw_heuristics import hrrn_policy, longest_queue_policy, shortest_queue_policy
from rw_heuristics import least_flexible_resource_policy, most_flexible_resource_policy
from rw_heuristics import least_flexible_activity_policy, most_flexible_activity_policy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import matplotlib.patches as mpatches
import os

problem_names = ['bpi2012', 'bpi2017', 'microsoft', 'production', 'consulta']
heuristics = [spt_policy, fifo_policy,
              hrrn_policy, longest_queue_policy, shortest_queue_policy,
              least_flexible_resource_policy, most_flexible_resource_policy,
              least_flexible_activity_policy, most_flexible_activity_policy]

# Create a mapping for more readable labels
heuristic_labels = {
    'ppo_policy': 'PPO',
    'ppo_policy_mid': 'PPO_mid',
    'spt_policy': 'SPT',
    'fifo_policy': 'FIFO',
    'random_policy': 'Random',
    'hrrn_policy': 'HRRN',
    'longest_queue_policy': 'LQF',
    'shortest_queue_policy': 'SQF',
    'least_flexible_resource_policy': 'LFRF',
    'most_flexible_resource_policy': 'MFRF',
    'least_flexible_activity_policy': 'LFAF',
    'most_flexible_activity_policy': 'MFAF'
}

# Iterate over each problem name
for problem_name in problem_names:
    data = []
    labels = []
    
    # Read data for each heuristic
    for heuristic in heuristic_labels.keys():
        try:
            df = pd.read_csv(f'results/{problem_name}/{problem_name}_{heuristic}.txt')
            cycle_times = df['cycle_time']
            data.append(cycle_times)
            labels.append(heuristic)
        except FileNotFoundError:
            print(f"File not found: results/{problem_name}/{problem_name}_{heuristic}.txt")
    
    if not data:  # Skip if no data found
        print(f"No data found for {problem_name}")
        continue
        
    # Create a box plot
    plt.figure(figsize=(12, 7))
    box = plt.boxplot(data, labels=[heuristic_labels.get(l, l) for l in labels], 
                     showmeans=True, patch_artist=True)
    
    # Calculate statistics
    mean_values = [np.mean(d) for d in data]
    std_values = [np.std(d) for d in data]
    
    # Find the best heuristic (lowest mean cycle time)
    best_policy_index = np.argmin(mean_values)
    best_policy = labels[best_policy_index]
    best_policy_data = data[best_policy_index]
    
    # Print mean and standard deviation for each policy
    print(f"\nMean and std for {problem_name}:")
    for label, mean, std in zip(labels, mean_values, std_values):
        print(f"{label}: {mean:.1f} ({std:.1f})")
    
    # Perform t-test between the best policy and all other policies
    p_values = []
    significance = []
    
    for i, (policy, policy_data) in enumerate(zip(labels, data)):
        if i == best_policy_index:
            p_values.append(1.0)  # Same distribution, p-value = 1
            significance.append(False)
        else:
            # Independent two-sample t-test
            stat, p = stats.ttest_ind(best_policy_data, policy_data, equal_var=True)
            p_values.append(p)
            significance.append(p < 0.05)  # Consider p < 0.05 as significant
    
    # Color the boxes based on significance
    for i, (patch, sig) in enumerate(zip(box['boxes'], significance)):
        if i == best_policy_index:
            patch.set_facecolor('lightgreen')  # Best policy in green
        elif sig:
            patch.set_facecolor('lightcoral')  # Significantly different in red
        else:
            patch.set_facecolor('lightblue')  # Not significantly different in blue
    
    # Add a legend
    best_patch = mpatches.Patch(color='lightgreen', label='Best policy')
    sig_patch = mpatches.Patch(color='lightcoral', label='Significantly different (p<0.05)')
    nonsig_patch = mpatches.Patch(color='lightblue', label='Not significantly different')
    plt.legend(handles=[best_patch, sig_patch, nonsig_patch], loc='upper right')
    
    # Add p-values above the boxes
    y_max = max([max(d) for d in data if len(d) > 0])
    y_min = min([min(d) for d in data if len(d) > 0])
    y_range = y_max - y_min
    for i, p in enumerate(p_values):
        if i != best_policy_index:
            plt.text(i+1, y_max + y_range * 0.05, f'p={p:.3f}', 
                     horizontalalignment='center', size='small')
    
    plt.title(f'Cycle Times for {problem_name}\nBest policy: {heuristic_labels.get(best_policy, best_policy)}')
    plt.ylabel('Cycle Time')
    plt.xlabel('Policy')
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Save the plot
    os.makedirs('figures', exist_ok=True)
    plt.savefig(f'figures/{problem_name}_boxplot.png')
    plt.show()
    
    # Print detailed test results
    print(f"\nSignificance test results for {problem_name}:")
    print(f"Best policy: {best_policy} (Mean: {mean_values[best_policy_index]:.2f})")
    print("-" * 50)
    for i, policy in enumerate(labels):
        if i != best_policy_index:
            print(f"{policy} vs {best_policy}: p-value = {p_values[i]:.5f} {'(significant)' if significance[i] else '(not significant)'}")
    print("\n")
```

```python
# In the second cell, update the table generation code:

# Add this code to create a LaTeX table
print("\\begin{table}[htbp]")
print("\\centering")
print("\\caption{Mean cycle times and standard deviations for different policies across problems}")
print("\\label{tab:policy_comparison}")
print("\\resizebox{\\textwidth}{!}{%")  # Make table fit page width

# Create a modified list with PPO and PPO_mid in the desired positions
table_heuristics = ['ppo_policy', 'ppo_policy_mid'] + [h.__name__ for h in heuristics]

# Start the tabular environment with the correct number of columns
# First column for problem names + one column per heuristic
headers = ["Event log"] + [heuristic_labels.get(h, h) for h in table_heuristics]
num_cols = len(headers)
print("\\begin{tabular}{l" + "r" * (num_cols-1) + "}")
print("\\toprule")

# Print the header row
print(" & ".join(headers) + " \\\\")
print("\\midrule")

# Store all results to create the table
all_results = {}

# Collect data for each problem
for problem_name in problem_names:
    data = []
    labels = []
    
    # Read data for each heuristic in our table order
    for heuristic_name in table_heuristics:
        try:
            df = pd.read_csv(f'results/{problem_name}/{problem_name}_{heuristic_name}.txt')
            cycle_times = df['cycle_time']
            data.append(cycle_times)
            labels.append(heuristic_name)
        except FileNotFoundError:
            data.append(np.array([]))
            labels.append(heuristic_name)
    
    if not any(len(d) > 0 for d in data):  # Skip if no data found
        continue
    
    # Calculate statistics for each policy
    mean_values = [np.mean(d) if len(d) > 0 else float('nan') for d in data]
    std_values = [np.std(d) if len(d) > 0 else float('nan') for d in data]
    
    # Find the best policy (lowest mean cycle time)
    valid_means = [m for m in mean_values if not np.isnan(m)]
    if valid_means:
        best_value = min(valid_means)
        best_indices = [i for i, m in enumerate(mean_values) if m == best_value]
    else:
        best_indices = []
    
    # Store results
    all_results[problem_name] = {
        'means': mean_values,
        'stds': std_values,
        'best_indices': best_indices
    }
    
    problem_names_formatted = {'bpi2012': 'BPI2012W',
                               'bpi2017': 'BPI2017W',
                               'microsoft': 'MC',
                               'production': 'PR',
                               'consulta': 'CDM'}

    # Create the row for this problem
    row_values = [problem_names_formatted[problem_name]]
    for i, (mean, std) in enumerate(zip(mean_values, std_values)):
        if np.isnan(mean):
            cell = ""  # Em dash for missing data
        else:
            # Bold the best value
            if i in best_indices:
                cell = f"\\textbf{{{mean:.1f}}}"
            else:
                cell = f"{mean:.1f}"
        row_values.append(cell)
    
    # Print the row
    print(" & ".join(row_values) + " \\\\")

print("\\bottomrule")
print("\\end{tabular}}")
print("\\caption*{Note: Values represent mean cycle time (standard deviation). Best policies for each problem are highlighted in bold.}")
print("\\end{table}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Define problem names (real-world processes)
problem_names = ['bpi2012', 'bpi2017', 'microsoft', 'production', 'consulta']

# Initialize a dictionary to store average action counts for each problem
action_counts_all = {
    'spt_policy': [],
    'fifo_policy': [],
    'hrrn_policy': [],
    'longest_queue_policy': [],
    'shortest_queue_policy': [],
    'least_flexible_resource_policy': [],
    'most_flexible_resource_policy': [],
    'least_flexible_activity_policy': [],
    'most_flexible_activity_policy': []
}

# More readable labels for the problems
problem_labels = {
    'bpi2012': 'BPI2012W',
    'bpi2017': 'BPI2017W',
    'microsoft': 'MC',
    'production': 'PR',
    'consulta': 'CDM'
}

# Iterate over each problem
for problem_name in problem_names:
    ppo_file_path = f'results/{problem_name}/{problem_name}_ppo_policy_mid.txt'
    if os.path.exists(ppo_file_path):
        ppo_df = pd.read_csv(ppo_file_path)
        
        # Calculate the mean of each action across all episodes
        # Only select columns that exist in the current DataFrame to avoid KeyError
        available_policies = [col for col in [
            'spt_policy', 'fifo_policy', 'hrrn_policy', 'longest_queue_policy',
            'shortest_queue_policy', 'least_flexible_resource_policy',
            'most_flexible_resource_policy', 'least_flexible_activity_policy',
            'most_flexible_activity_policy'
        ] if col in ppo_df.columns]
        
        action_means = ppo_df[available_policies].mean()
        
        # Calculate total actions for this problem
        total = action_means.sum()
        
        # Append the percentage of each action to the dictionary
        for policy in action_means.index:
            action_counts_all[policy].append((action_means[policy] / total) * 100)
    else:
        print(f"File not found: {ppo_file_path}")
        # Append zeros if the file doesn't exist
        for policy in action_counts_all.keys():
            action_counts_all[policy].append(0)

# Create a stacked bar chart
plt.figure(figsize=(7,4.2), dpi=300)  # Higher resolution, better aspect ratio

# Use a professional, colorblind-friendly palette
colors = sns.color_palette('deep', 9)
bottom = np.zeros(len(problem_names))

# Better labels for the legend
action_labels = {
    'spt_policy': 'SPT',
    'fifo_policy': 'FIFO',
    'hrrn_policy': 'HRRN',
    'longest_queue_policy': 'LQF',
    'shortest_queue_policy': 'SQF',
    'least_flexible_resource_policy': 'LFRF',
    'most_flexible_resource_policy': 'MFRF',
    'least_flexible_activity_policy': 'LFAF',
    'most_flexible_activity_policy': 'MFAF'
}

# Add subtle grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.8, zorder=0)

# Initialize bars dictionary to store the bar plot objects
bars = {}

# Plot each policy with improved styling
for i, (policy, percentages) in enumerate(action_counts_all.items()):
    if any(percentages):  # Only plot policies that were used
        bars[policy] = plt.bar([problem_labels.get(problem, problem) for problem in problem_names],
                               percentages, bottom=bottom, label=action_labels[policy],
                               color=colors[i], edgecolor='white', linewidth=0.5,
                               zorder=3)
        
        # Add percentage labels for segments larger than 5%
        for j, (pct, btm) in enumerate(zip(percentages, bottom)):
            if pct > 5:  # Only show label if percentage is large enough
                plt.text(j, btm + pct/2, f'{pct:.1f}%', 
                         ha='center', va='center', fontsize=9, 
                         color='white', fontweight='bold')
        
        bottom += np.array(percentages)

# Improve axes and labels
plt.ylabel('Percentage of Actions (%)', fontsize=12, fontweight='bold')
plt.xlabel('Event Log', fontsize=12, fontweight='bold')

# Improve tick parameters
plt.yticks(fontsize=10)

# Remove top and right spines for cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Better legend placement and styling
plt.legend(fontsize=10, bbox_to_anchor=(0.5, 1.00), loc='lower center',
           framealpha=0.9, edgecolor='lightgray', ncol=5)

plt.ylim(0, 100)
plt.tight_layout()

# Create figures directory if it doesn't exist
os.makedirs('figures', exist_ok=True)

# Save in high resolution for publication
plt.savefig('figures/action_distribution_real_world_mid.pdf', bbox_inches='tight', dpi=300)
#plt.savefig('figures/action_distribution_real_world.png', bbox_inches='tight', dpi=300)

plt.show()
```
